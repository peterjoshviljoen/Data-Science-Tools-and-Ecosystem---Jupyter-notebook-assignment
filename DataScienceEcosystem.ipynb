There is an enormous amount of programming languages that have potential use cases in the field of data science, many of which have their own weaknesses or strengths, but this large assortment can be daunting for someone who is a beginner wanting to engage with the field. One can look to some of the most commonly used programming languages to start, these include the following:

1. Python - probably the most popular and widely used programming language in the data science field. Python is an open source programming language and has seen use in a wide varity of areas such data analysis, data visualization, AI and machine learning.
2. R - is free software that is popular amongst statisticians, mathematicians and data miners who use it to develope statistical software, data analysis and visualization. R has an array-orientated syntax which makes it easier to translate from mathematics to code, which may be ideal who those with little programming background or knowledge.
3. SQL - or Structured Query Language, is a non-procedutral programming language limited to querying and managaing the data of relational databases.
4. C++ - is an extension of the C programming language and is a good general purpose programming language. TensorFlow, which is a widely used Deep Learning library, at its foundation was built using C++ even though it runs on a Python interface.


It is actually possible to extend the functionalites of these languages using what we call "libraries", and many of these libraries come with very handy tools for data scientists. I had brifly mentioned in the list above that TensorFlow (used for developing deep learning models) is a library that is built with C++. That is one of many indispensable libraries that data scientists have in their toolkit. Here are a few other examples of common libraries used for data science:

1. NumPy - these libraries are based upon arrays and matrices and has functions that are useful in the domain of linear algebra and fourier transforms.
2. Matplotlib - this library is very well known for its data visualization capabilities. It is popular for the ease of creating and customizability of its graphs/plots in Python.
3. ggplot2 - this is a popular library used for data visualization in R.
4. Scikit-learn - is a library that contains tools for developing machine learning models, including regression, classification, clustering and so on. It was built upon NumPy, SciPy and Matplotlib.
